Cargamos el CSV

```python
import pandas as pd

#Tren
df = pd.read_csv('CoordenadasMaritimoEntrenamientoSucias.csv', usecols=['punto_x', 'lat_x', 'long_x', 'punto_y', 'lat_y', 'long_y', 'distance', 'speedKmH', 'Climate', 'eta', 'Date', "eta/km*100", "eta/km"], sep=",")

#Barco
#df = pd.read_csv('CoordenadasTrenEntrenamientoSucias.csv', usecols=['punto_x', 'lat_x', 'long_x', 'punto_y', 'lat_y', 'long_y', 'distance', 'speedKn', 'speedKmH', 'Climate', 'eta', 'Date', "eta/km*100", "eta/km"], sep=",")
```

Obtenemos el mes de las fechas porque es lo que realmente nos interesa

```python
df['month'] = pd.DatetimeIndex(df['Date']).month #Obtenemos los meses
```

Visualizamos los datos en función del mes y el clima

```python
import seaborn as sns
import matplotlib.pyplot as plt

grafico = sns.stripplot(data=df, x="month", y="eta/km*100", palette="deep", hue="Climate", dodge="True", size=5)
plt.setp(grafico.get_legend().get_texts(), fontsize='8') #Cambiamos tamaño de Fuente en leyenda
grafico.legend(loc='upper right', ncol=3 ) #Situamos la leyenda y la dividimos en columnas
plt.show()
```

Obtenemos las medias

```python
#Primero la media de todos los etas
print(df['eta/km*100'].mean())

#La media del eta dividido por cada clima y mes
print(df.groupby(['Climate', 'month'])['eta/km*100'].mean().to_string()) #Con el .toString() forzamos que se imprima entero
```

Desviaciones

```python
import numpy as np

df_eta = df['eta/km*100']
x = np.std(df_eta)
print('Desviacion total = '  + str(x))

#Desviacion por clima
print(df.groupby(['Climate'])['eta/km*100'].std())

#Sigue siendo alta

#Desviacion por mes
print(df.groupby(['month'])['eta/km*100'].std())

#Tambien alta en algunos meses

#Desviacion por clima y mes
print(df.groupby(['Climate', 'month'])['eta/km*100'].std().to_string())

#En algunos casos sigue siendo alta, pero en la mayoría ha bajado, sin embargo, habría que hacer 6x12 = 72 limpiezas

#Se observa que generalmente cada 3 meses la desviación es similar dentro del mismo clima, por lo que la limpieza de casos extremos se puede hacer 
#operando conjuntamente cada 3 meses dentro del mismo clima.


Climate = ['Sunny', 'Cloudy', 'Foggy', 'Rain', 'Storm', 'Hurricane']
for cl in Climate:
    print (cl)
    for num in range(4, 14, 3):
        print ("Mes: " + str(num-3) + " - " + str(num-1))
        df_Cl_1 = df.loc[(df['Climate'] == cl) & (df['month'] < num)]
        std = df_Cl_1.groupby(['month'])['eta/km*100'].std().std() #Obtenemos la desviacion de las desviaciones cada 3 meses y vemos que es muy baja, por lo que se puede trabajar con esos datos grupalmente
        print(std)
        

```

Percentiles/Casos Extremos

```python


for cl in Climate:
    print (cl)
    for num in range(4, 14, 3):
        print (num-3, "-", num-1 )
        df_Cl = df.loc[(df['Climate'] == cl) & (df['month'] < num)]
        print('Valor Máximo: ')
        print(df_Cl['eta/km*100'].max())
        print('Media: ')
        print(df_Cl['eta/km*100'].mean())
        print('Mediana: ')
        print(df_Cl['eta/km*100'].median())
        print('Desviacion: ')
        print(df_Cl['eta/km*100'].std())     
        print('Percentil 95')
        print(df_Cl['eta/km*100'].quantile(0.95))
        print('Percentil 99')
        print(df_Cl['eta/km*100'].quantile(0.99))
        print('Percentil 99.5')
        print(df_Cl['eta/km*100'].quantile(0.995))
        print('Percentil 99.9')
        print(df_Cl['eta/km*100'].quantile(0.999))
        print('Percentil 99.95')
        print(df_Cl['eta/km*100'].quantile(0.9995))
        #Casos extremos eta bajo
        print('Percentil 5')
        print(df_Cl['eta/km*100'].quantile(0.05))
        print('Percentil 1')
        print(df_Cl['eta/km*100'].quantile(0.01))
        print('Percentil 0.5')
        print(df_Cl['eta/km*100'].quantile(0.005))
        print('Percentil 0.1')
        print(df_Cl['eta/km*100'].quantile(0.001))
        print('Percentil 0.05')
        print(df_Cl['eta/km*100'].quantile(0.0005))
         



```

Observamos que por debajo del percentil 0.1, hay un salto bastante grande en los valores que no se corresponde con la progresión en los percentiles previos.

Realizamos la limpieza de estos datos y observamos el resultado del conjunto de datos

```python
L = []
for cl in Climate:
    print (cl)
    for num in range(4, 14, 3):
        print (num)
        df_Temp = df.loc[(df['Climate'] == cl) & (df['month'] < num)]
        bajo = df_Temp['eta/km*100'].quantile(0.05)
        alto = df_Temp['eta/km*100'].quantile(0.95)
        df_Temp = df_Temp[df_Temp['eta/km*100'] > bajo]
        df_Temp = df_Temp[df_Temp['eta/km*100'] < alto]
        L.append(df_Temp)
        
out = pd.concat(L)
```

Observamos que por debajo del percentil 0.1, hay un salto bastante grande en los valores que no se corresponde con la progresión en los percentiles previos.
Lo mismo ocurre en el percentil 99.9, por esto deducimos que el comportamiento típico se encuentra entre el percentil 0.5 y 99.5, y ya que este sistema no pretende predecir
casos extremos.

Realizamos la limpieza de estos datos y observamos el resultado del conjunto de datos
