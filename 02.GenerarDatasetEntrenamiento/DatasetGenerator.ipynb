Dataset Generator

```python
import pandas as pd
import networkx as nx
import math

#df = pd.read_csv('Tren_Y_Barco.csv', usecols=['punto', 'x', 'y', 'modo'], sep=";")
df = pd.read_csv('CoordenadasMaritimoLimpias.csv', usecols=['punto', 'lat', 'long', 'modo'], sep=";")


df_cp = df #Para devolver los puntos al csv posteriormente

df = df.merge(df, how="cross") #Join de cada punto con los demás
df.head
df = df.loc[df.apply(lambda x: x["punto_x"] != x["punto_y"] or not any((df["punto_x"] == x["punto_y"]) & (df["punto_y"] == x["punto_x"])), axis=1)]

 #Eliminamos conexion de puntos consigo mismos

df["distance"] = df.apply(lambda x: (math.hypot(x["lat_x"]-x["lat_y"], x["long_x"]-x["long_y"]) * 100), axis=1) #Obtenemos la distancia entre los puntos Hipotenusa de triángulo formado por dist X y dist 


df = df.reset_index(drop=True)

print(df)

#df.to_csv("dfPruebaWeatherDist.csv", index=False) #Para generar el dataset posteriormente
```

Calculamos velocidad. TREN

```python
import pandas as pd
import random 

#Media en España de mercancias es 56 km/h
df["speedKmH"] = df.apply(lambda x: random.uniform(40,70), axis=1) #Obtenemos un valor aleatorio en todas las columnas de velocidad en nudos
```

Calculamos velocidad. MARITIMO

```python
import pandas as pd
import random 

df["speedKn"] = df.apply(lambda x: random.uniform(7,15), axis=1) #Obtenemos un valor aleatorio en todas las columnas de velocidad en nudos
df["speedKmH"] = df['speedKn'] * 1.852 #Pasamos los nudos a km/h
```

```python
import pandas as pd
import random 

climate = ["Sunny", "Cloudy", "Rain", "Storm", "Foggy", "Hurricane"] #Climas posibles

df["Climate"] = df.apply(lambda x: random.choice(climate), axis=1)


df["eta"] = df["distance"]/df["speedKmH"] * random.uniform(0.8,1.2) #Error de 0.8 a 1.2


df.head(20)

#df.to_csv("prueba00.csv", index=False)


```

Aplicamos un facor correctivo al ETA en función del clima

```python
import numpy as np

min_max = {'Sunny': (0.8, 1.0), 'Rain': (1.1, 1.4), 'Cloudy': (1.0, 1.2), 'Foggy': (1.2, 1.6),  'Storm': (1.4, 1.8),  'Hurricane': (1.6, 2.2)}

df['eta'] = (df.groupby('Climate')['eta']
               .apply(lambda x: x*np.random.uniform(*min_max[x.name], size=len(x))) #Que hace el *antes de min_max
             )

df.head(20)
```

Generar Fecha

```python
import time
from datetime import date
import random

def random_date(seed):
    random.seed(seed)
    d = random.randint(1, int(time.time()))
    return date.fromtimestamp(d).strftime('%Y-%m-%d')

print(random_date(random.uniform(0,100000))) #semilla aleatoria

df["Date"] = df.apply(lambda x: random_date(random.uniform(0,100000)), axis=1)

df.head(10)



```

Aumentar Tiempo dependiendo de fechas

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

#Multiplica todos por el mismo
#df.loc[df['Date'].dt.month.isin((1,2,3)), 'eta'] *= random.uniform(1,2)

#Escogemos meses

#Primavera
m = df['Date'].dt.month.isin((4,5,6))

#Será buen clima
df.loc[m, 'eta'] *= np.random.uniform(0.8, 1.0, m.sum()) #Multiplicamos por rango aleatorio

#Verano
m = df['Date'].dt.month.isin((7,8,9))

#Será buen clima pero el excesivo calor puede suponer retrasos
df.loc[m, 'eta'] *= np.random.uniform(0.9, 1.1, m.sum()) #Multiplicamos por rango aleatorio

#Otoño
m = df['Date'].dt.month.isin((10,11,12))

#Habrá más lluvias
df.loc[m, 'eta'] *= np.random.uniform(1.2, 1.4, m.sum()) #Multiplicamos por rango aleatorio

#Invierno
m = df['Date'].dt.month.isin((1,2,3))

#Más frío más complicaciones, peor situación.
df.loc[m, 'eta'] *= np.random.uniform(1.4, 1.6, m.sum()) #Multiplicamos por rango aleatorio


df.head()

```

Introducimos casos extremos

```python
numExtreme = 200 #Amount of registres to apply the random modification.


mod_indices_alto = np.random.choice(df.index, numExtreme, replace=False)
print(mod_indices_alto)
alto = np.random.uniform(1.8, 2.1, len(mod_indices_alto)) #Casos extremos, eta alto
df.loc[mod_indices_alto, 'eta'] *= alto

mod_indices_bajo = np.random.choice(df.index, numExtreme, replace=False)
print(mod_indices_bajo)
bajo = np.random.uniform(0.2, 0.4, len(mod_indices_bajo)) #Casos extremos, eta bajo
df.loc[mod_indices_bajo, 'eta'] *= bajo
```

Obtenemos eta/km*100 para analisis y limpieza del dataset posteriormente y eta/km para calculo de prediccion del eta

```python
df["eta/km*100"] = df["eta"]/df["distance"]*100 #Multiplicamos x 100 para tener valores más fácilmente visualizables (Horas) De cara a analisis
df["eta/km"] = df["eta"]/df["distance"] #Se usará para predecir 

df.to_csv("CoordenadasMaritimoEntrenamientoSucias.csv")
```
